```python
#Qno01
def string_operations(s):
    return s[::-1], s[::2]

print(string_operations("hello"))
```

```python
#Qno02
def squares_of_evens(numbers):
    return [n**2 for n in numbers if n % 2 == 0]

print(squares_of_evens([1,2,3,4,5,6]))
```

```python
#Qno03
def char_frequency(s):
    freq = {}
    for ch in s:
        freq[ch] = freq.get(ch, 0) + 1
    return freq

print(char_frequency("hello"))
```

```python
#Qno04
def odd_indices(lst):
    return [index for index, value in enumerate(lst) if value % 2 != 0]

print(odd_indices([1,2,3,4,5,6]))
```

```python
#Qno05
def rotate_right(lst, k):
    k = k % len(lst)
    return lst[-k:] + lst[:-k]

print(rotate_right([1,2,3,4,5], 2))
```

```python
#Qno06
def lists_to_dict(keys, values):
    return dict(zip(keys, values))

print(lists_to_dict(["a","b","c"], [1,2,3]))
```

```python
#Qno07
def manual_iteration(lst):
    it = iter(lst)
    while True:
        try:
            print(next(it))
        except StopIteration:
            break

manual_iteration([10,20,30])
```

```python
#Qno08
def flatten_list(nested_list):
    return [item for sublist in nested_list for item in sublist]

print(flatten_list([[1,2],[3,4],[5,6]]))
```

```python
#Qno09
def is_palindrome(s):
    return s == s[::-1]

print(is_palindrome("madam"))
print(is_palindrome("hello"))
```

```python
#Qno10
def second_largest(lst):
    largest = second = float('-inf')
    for num in lst:
        if num > largest:
            second = largest
            largest = num
        elif largest > num > second:
            second = num
    return second

print(second_largest([10, 20, 4, 45, 99]))
```
